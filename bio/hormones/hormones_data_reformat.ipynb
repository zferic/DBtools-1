### reformat rules:
### "QNS" => -999, "use -998, QNS=quantity not sufficient for testing"
### "na" => -999, "use -999, na=no result available or returned"
### ">15,000" => 15000, add "real value is >15,000" to note section


```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('Georgetown_Univ_Hormones_export_5.11.16_neu.csv')
df
```

```python
list(df.columns.values)
```

```python
df.loc[df.conc == 'na', 'notes '] = 'use -999, na=no result available or returned'
df.loc[df.conc == 'na', 'conc'] = -999
```

```python
df.loc[df.conc == 'QNS', 'notes '] = 'use -998, QNS=quantity not sufficient for testing'
df.loc[df.conc == 'QNS', 'conc'] = -998
```

```python
df.loc[df.conc == '> 15,000', 'notes '] = 'use 15000, the actual value is >15,000'
df.loc[df.conc == '> 15,000', 'conc'] = 15000
## some variations
df.loc[df.conc == '>15,000', 'notes '] = 'use 15000, the actual value is >15,000'
df.loc[df.conc == '>15,000', 'conc'] = 15000

df.loc[df.conc == '>15000', 'notes '] = 'use 15000, the actual value is >15,000'
df.loc[df.conc == '>15000', 'conc'] = 15000
```

```python
df.to_csv('test.csv')
```
